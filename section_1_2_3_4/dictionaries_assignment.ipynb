```python
#11
dic1 = {i:i**2 for i in range(1,11)}
dic2 = {i:dic1[i] for i in dic1 if i%2==0}
print(dic1)
print(dic2)

#12
dic1 = {i:i**2 for i in range(1,11)}
dic2 = {dic1[i]:i for i in dic1 if i%2!=0}
print(dic1)
print(dic2)

#13
from collections import defaultdict

default_dict = defaultdict(list)
default_dict['a'].append(1)
default_dict['a'].append(2)
default_dict['b'].append(3)
print(default_dict)

#14
a = input("Enter a string: ")
dic = {i:a.count(i) for i in a}
print(dic)

#15
import json

book = { "title": "Python", "author": "Guido", "year": 2023, "genere": "programming"}
book_json = json.dumps(book, indent=4)
print(book_json)
```
