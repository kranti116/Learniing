```python
#1
def febonacci(n):
    # print("next number:",n)
    if n < 2:
        return n
    return febonacci(n-1) + febonacci(n-2)

print(febonacci(10))

#2
def createDic(a, b={}):
    b.update({a: a**2})
    return b

print(createDic(2))
print(createDic(3, {1:1}))

#3
def varKeys(*a):
    return {i:int(i)**2 for i in a if i.isdigit()}
print(varKeys("1","2","three","4"))

def varkeyword(**a):
    return {i:int(j)**2 for i,j in a.items() if j.isdigit()}

print(varkeyword(a="1", b="2", c="three", d="4"))

#4 #5
def anotherdef(x):
    return x**3

def callback_fun(anotherdef, lst):
    return [anotherdef(x) for x in lst]

print(callback_fun(anotherdef, [1,2,3,4]))

#6
import time
def clock_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Time taken is {end_time - start_time}")
        return result
    return wrapper

@clock_time
def complex_task(lst):
    return {i:i**2 for i in lst}

print(complex_task([1,2,3]))
```

```python
#7
# Define the apply_function
def apply_function(filter_func, map_func, lst):
    return [map_func(x) for x in lst if filter_func(x)]

# Example list
z = [1, 'a', 3, 'b', 5]

# Call apply_function with the specified arguments
result = apply_function(
    lambda x: isinstance(x, int),  # Filter function to filter integers
    lambda x: x * 2,               # Map function to multiply by 2
    z                              # List of numbers and characters
)

print(result)  # Output: [2, 6, 10]


#8
def func_1(list):
    print('in function1',list)
    y = func2(list)
    return {x:y[x]+3 for x in y.keys()}

def func2(a):
    print('in func2')
    b =  {'a'+str(x):x**2 for x in a}
    print('func2',b)
    return b

def main_func(list):
    print('this is list', list)
    return func_1(list)

print(main_func([1,2,3]))

#9
from functools import partial

multiply_by_2 = partial(lambda x, y: x + y * x, 3)

# Test
print(multiply_by_2(3))  # 6
print(multiply_by_2(5))  # 10

#10
def avgfunc(lst):
    if lst:
        return sum(lst) / len(lst)
    else:
        return None
    
print(avgfunc([1,2,3]))
print(avgfunc([]))
a = None
print(avgfunc(a))
```

```python
#11
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b


fib_gen = fibonacci_generator()
for _ in range(10):
    print(next(fib_gen))

#12
```
